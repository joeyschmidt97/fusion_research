```python
def plot_limit(x, y, limit, perc = .1):
    dy = np.diff(y)
    
    dy_rescaled = dy/max(dy)
    # limit = 1e-3
    pass_limit = abs(dy_rescaled) > limit

    global front_pass_limit, stop, start
    flip_switch = True
    for i,item in enumerate(pass_limit):
        if item and flip_switch:
            start = i
            front_pass_limit = pass_limit[i:]
            flip_switch = False
    

    rev_pass_limit = np.flip(front_pass_limit)
    # rev_pass_limit = front_pass_limit[::-1]

    flip_switch = True
    for i,item in enumerate(rev_pass_limit):
        if item:
            stop = i
            flip_switch = False
    
    start -= round(perc*len(x))
    stop += round(perc*len(x))

    x_start = x[start]
    x_stop = x[-stop]
    
    x_lim = max(abs(x_start), abs(x_stop))
    
    return x_lim
```

```python
x_lim = plot_limit(x,y, 1e-3, .0001)

plt.plot(x,y)
plt.xlim(-x_lim, x_lim)
plt.show()


# print(x[:])
# print(x[:-1])

print(min(-11,abs(-2),3))
```
