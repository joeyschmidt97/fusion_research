```python
import EV_kHz_DF
import Omega_kHz_DF

import matplotlib.pyplot as plt
```

### Load EV Data

```python
efit2_EV_path = "/global/cscratch1/sd/joeschm/AUG/efit2_EVS"
miller2_EV_path = "/global/cscratch1/sd/joeschm/AUG/miller2_EVS"

efit2_EV_omega = EV_kHz_DF.omega_EV_df(efit2_EV_path).reset_index()
miller2_EV_omega = EV_kHz_DF.omega_EV_df(miller2_EV_path).reset_index()

#print(miller2_EV_omega)
```

### Load IV data

```python
efit2_path = "/global/cscratch1/sd/joeschm/AUG/efit2"
miller2_path = "/global/cscratch1/sd/joeschm/AUG/miller2"

efit2_omega = Omega_kHz_DF.omega_df(efit2_path)
miller2_omega = Omega_kHz_DF.omega_df(miller2_path)
```

### Plot IV Data (efit2 vs miller2)

```python
ax = efit2_omega.plot(x ='n0_global', y='omega(kHz)', kind = 'scatter', color = 'red', marker = '+',s=120, label = 'efit2', alpha = 0.5)
miller2_omega.plot(x ='n0_global', y='omega(kHz)', kind = 'scatter',color = 'blue', marker = '+', s=120,  ax = ax, label = 'miller2')

#plt.ylim(-30000,-3000)
plt.legend()
plt.show()
```

### Plot EV Data (efit2 vs miller2)

```python
ax = efit2_EV_omega.plot(x ='n0_global', y='omega(kHz)', kind = 'scatter', color = 'red', label = 'efit2', alpha = 0.2)
miller2_EV_omega.plot(x ='n0_global', y='omega(kHz)', kind = 'scatter', ax = ax, label = 'miller2', alpha = 0.2)

#plt.ylim(-30000,-3000)
plt.legend()
plt.show()
```

### Plot miller2 Data (EV vs IV)

```python
ax = miller2_omega.plot(x ='n0_global', y='omega(kHz)', kind = 'scatter', color = 'blue', marker = '+', s=180, label = 'miller2_IV')
miller2_EV_omega.plot(x ='n0_global', y='omega(kHz)', kind = 'scatter', ax = ax, color = 'blue', label = 'miller2_EV', alpha = 0.2)

#plt.ylim(-30000,-3000)
plt.legend()
plt.show()
```

### Plot efit2 Data (EV vs IV)

```python
ax = efit2_omega.plot(x ='n0_global', y='omega(kHz)', kind = 'scatter', color = 'red', marker = '+',s=180, label = 'efit2_IV', alpha = 0.5)
efit2_EV_omega.plot(x ='n0_global', y='omega(kHz)', kind = 'scatter', ax = ax, color = 'red', label = 'efit2_EV', alpha = 0.2)

#plt.ylim(-30000,-3000)
plt.legend()
plt.show()
```
