## EFIT vs Miller Plots (DIIID)

```python
efit_path = "/global/cscratch1/sd/joeschm/162940/efit"
miller_path = "/global/cscratch1/sd/joeschm/162940/miller"

efit_omega = om.omega_df(efit_path, 500, 600)
miller_omega = om.omega_df(miller_path, 500, 600)

efit_omega_2x = om.omega_df(efit_path, 600, 1300)
miller_omega_2x = om.omega_df(miller_path, 600, 1300)

#print(efit_omega)
#print(miller_omega_2x)
```
