## Miller on Miller & Miller 2 Geometry

```python
m_path = "/global/cscratch1/sd/joeschm/162940/miller"
m_omega = omDF.omega_df(m_path, 500, 600)
m_omega_low = omDF.omega_df(m_path, 200, 500)

m_m2_geom_path = "/global/cscratch1/sd/joeschm/162940/miller_miller2_geom"
m_m2_geom_omega = omDF.omega_df(m_m2_geom_path, 500, 600)
m_m2_geom_omega_low = omDF.omega_df(m_m2_geom_path, 200, 500)


#print(m_omega_low) #longer runs (or more cores) for miller on miller geom
#print(m_m2_geom_omega_low)
```

```python
ax = m_omega.plot(x ='n0_global', y='omega(kHz)', kind = 'scatter', marker = '+', s = 180, color = 'red', label = 'miller (M_geom) [512]', alpha = 0.5)
m_m2_geom_omega.plot(x ='n0_global', y='omega(kHz)', kind = 'scatter', ax = ax, label = 'miller (M2_geom) [512]',marker = '+', s = 180, color = 'blue',  alpha = 0.5)

m_omega_low.plot(x ='n0_global', y='omega(kHz)', kind = 'scatter', ax = ax, label = 'miller (M_geom) [256]', color = 'red',  alpha = 0.5)
m_m2_geom_omega_low.plot(x ='n0_global', y='omega(kHz)', kind = 'scatter', ax = ax, label = 'miller (M2_geom) [256]', color = 'blue',  alpha = 0.5)


#plt.ylim(-30000,-3000)
plt.legend()
plt.show()
```
