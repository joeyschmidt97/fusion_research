```python
import sys
sys.path.insert(0, '/global/homes/j/joeschm/Notebooks/omega_reading')
import om_kHz_DF_V2 as om
import matplotlib.pyplot as plt
```

```python
path_V0_efit = '/global/cscratch1/sd/joeschm/162940/efit'
path_V0_miller = '/global/cscratch1/sd/joeschm/162940/miller'

path_V1_efit = '/global/cscratch1/sd/joeschm/162940/COMP_efit'
path_V1_miller = '/global/cscratch1/sd/joeschm/162940/COMP_miller'

path_V2_efit = '/global/cscratch1/sd/joeschm/162940/V2_efit'
path_V2_miller = '/global/cscratch1/sd/joeschm/162940/V2_miller'


V0_efit_omega = om.omega_df(path_V0_efit, [], [])
V0_miller_omega = om.omega_df(path_V0_miller, [], [])

V1_efit_omega = om.omega_df(path_V1_efit, [], [])
V1_miller_omega = om.omega_df(path_V1_miller, [], [])

V2_efit_omega_512 = om.omega_df(path_V2_efit, ['nz0 == 512'], [])
V2_miller_omega_512 = om.omega_df(path_V2_miller, ['nz0 == 512'], [])
```

```python
size = 100
marker_style = '.'

ax = V0_efit_omega.plot(x ='n0_global', y='omega(kHz)', kind = 'scatter', color = 'red', label = 'V0_efit', alpha = 0.5, marker = marker_style, s = size, figsize=(10,10))
V0_miller_omega.plot(x ='n0_global', y='omega(kHz)', kind = 'scatter', color = 'blue', label = 'V0_miller',  alpha = 0.5, marker = marker_style, s=size, ax = ax)

V1_efit_omega.plot(x ='n0_global', y='omega(kHz)', kind = 'scatter', color = 'red', label = 'V1_efit',  alpha = 0.5, marker = '>', s=size, ax = ax)
V1_miller_omega.plot(x ='n0_global', y='omega(kHz)', kind = 'scatter', color = 'blue', label = 'V1_miller',  alpha = 0.5, marker = '<', s=size, ax = ax)

V2_efit_omega_512.plot(x ='n0_global', y='omega(kHz)', kind = 'scatter', color = 'red', label = 'V2_efit (512)', alpha = 0.5, marker = '+', s = 200, ax=ax)
V2_miller_omega_512.plot(x ='n0_global', y='omega(kHz)', kind = 'scatter', color = 'blue', label = 'V2_miller (512)',  alpha = 0.5, marker = '+', s= 200, ax = ax)

plt.legend()
plt.show()


ax = V0_efit_omega.plot(x ='n0_global', y='gamma(kHz)', kind = 'scatter', color = 'red', label = 'V0_efit', alpha = 0.5, marker = marker_style, s = size, figsize=(10,10))
V0_miller_omega.plot(x ='n0_global', y='gamma(kHz)', kind = 'scatter', color = 'blue', label = 'V0_miller',  alpha = 0.5, marker = marker_style, s=size, ax = ax)

V1_efit_omega.plot(x ='n0_global', y='gamma(kHz)', kind = 'scatter', color = 'red', label = 'V1_efit',  alpha = 0.5, marker = '>', s=size, ax = ax)
V1_miller_omega.plot(x ='n0_global', y='gamma(kHz)', kind = 'scatter', color = 'blue', label = 'V1_miller',  alpha = 0.5, marker = '<', s=size, ax = ax)

V2_efit_omega_512.plot(x ='n0_global', y='gamma(kHz)', kind = 'scatter', color = 'red', label = 'V2_efit (512)', alpha = 0.5, marker = '+', s = 200, ax=ax)
V2_miller_omega_512.plot(x ='n0_global', y='gamma(kHz)', kind = 'scatter', color = 'blue', label = 'V2_miller (512)',  alpha = 0.5, marker = '+', s=200, ax = ax)

plt.legend()
plt.show()
```
