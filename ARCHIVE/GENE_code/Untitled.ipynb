```python
def plot_points(ref_x_list, ref_y_list, x_list, y_list):
    plt.figure(figsize=(10, 6))

    # Plot reference points
    plt.plot(ref_x_list, ref_y_list, 'bo-', label='Reference Points')

    # Plot other points
    plt.plot(x_list, y_list, 'ro-', label='Other Points')

    # Plot percent difference as a bar plot
    # plt.bar(x_list, percent_diff_values, alpha=0.5, width=0.3, align='center', label='Percent Difference')

    plt.xlabel('X')
    plt.ylabel('Y')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
def mod_plot_points(ref_x_list, ref_y_list, x_list, y_list, low_bound):
    
    x_combined = ref_x_list + x_list
    x_combined = [*set(x_combined)]
    x_combined = sorted(x_combined)
    
    MOD_ref_y_list = interpolate_points(x_combined, ref_x_list, ref_y_list)
    MOD_y_list = interpolate_points(x_combined, x_list, y_list)
    
    fig, (ax, ax_diff) = plt.subplots(nrows=2, gridspec_kw={'height_ratios': [2, 1]}, figsize=(5, 6))
    # plt.figure(figsize=(10, 6))
    
    # Plot reference points
    # plt.plot(x_combined, MOD_ref_y_list, 'bo-', label='Reference Points')

    # Plot other points
    # plt.plot(x_combined, MOD_y_list, 'ro-', label='Other Points')
    
#     ax.plot(x_combined, MOD_ref_y_list, 'bo-', label='Reference Points')
#     ax.plot(x_combined, MOD_y_list, 'ro-', label='Other Points')
    
    # Plot reference points
    ax.plot(ref_x_list, ref_y_list, 'bo-', label='Reference Points')
    ax.plot(x_list, y_list, 'ro-', label='Other Points')
    
    ax.set_xlabel('X', fontsize=15)
    ax.set_ylabel('Y', fontsize=15)
    ax.legend()
 
    percent_diff_values = []
    for val_ref, val in zip(MOD_ref_y_list, MOD_y_list):
        
        # low_bound = 1
        
        if (val_ref < 0) or (val < 0):
            percent_diff = 100*(val_ref - val / abs(val_ref))
        
        else:
            percent_diff = 100*abs(val_ref - val) / ((val_ref + val)/2)
        
        percent_diff_values.append(percent_diff)
        
#         percent_diff = 100*(val_ref - val) / (abs(val_ref))
#         percent_diff_values.append(percent_diff)
    
    # print(percent_diff_values)
    

    ax_diff.set_ylabel('% Diff', fontsize=15)
    
    
    ax_diff.set_xlabel(low_bound, fontsize=15)
    ax_diff.plot(x_combined, percent_diff_values, linestyle='dotted', marker='o', color='green')
    
    
```

```python
def interpolate_points(X_MERGE, x_list, y_list):
    # print(X_MERGE)
    # print(x_list, y_list)
    
    Y_MERGE = []
    
    for i, X in enumerate(X_MERGE):
        closest_x = min(x_list, key=lambda x:abs(x - X))
        closest_x_i = x_list.index(closest_x)
        
        delta_x = X - closest_x
        
        in_between = True
        
        if (X == closest_x):
            in_between = False
            Y_int = y_list[closest_x_i]
        
        elif (X < x_list[0]):
            in_between = False
            x_back = x_list[0]
            y_back = y_list[0]
            x_front = x_list[1]
            y_front = y_list[1]
            
            m = (y_front - y_back)/(x_front - x_back)
            Y_int = m*delta_x + y_back
            
        elif (X > x_list[-1]):
            in_between = False
            x_back = x_list[-2]
            y_back = y_list[-2]
            x_front = x_list[-1]
            y_front = y_list[-1]
            
            m = (y_front - y_back)/(x_front - x_back)
            Y_int = m*delta_x + y_front
            
        elif (delta_x < 0) and in_between:       
            # print('left')
            
            x_back = x_list[closest_x_i - 1]
            y_back = y_list[closest_x_i - 1]
            x_front = x_list[closest_x_i]
            y_front = y_list[closest_x_i]
            
            m = (y_front - y_back)/(x_front - x_back)
            Y_int = m*delta_x + y_front
            
        elif (delta_x > 0) and in_between:
            # print('right')
            
            x_back = x_list[closest_x_i]
            y_back = y_list[closest_x_i]
            x_front = x_list[closest_x_i + 1]
            y_front = y_list[closest_x_i + 1]
            
            m = (y_front - y_back)/(x_front - x_back)
            Y_int = m*delta_x + y_back
        
        Y_MERGE.append(Y_int)
    
    
    return Y_MERGE
```
